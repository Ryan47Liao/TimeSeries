```python
import pandas as pd
import matplotlib.pyplot as plt
import os
import datetime
```

```python
path = 'E:\Data\TS_hw5'
DATABASE = dict()
for name in os.listdir(path):
    print(name)
    file_path = path + '/' + name
    DATABASE[name.split('.')[0]] = pd.read_csv(file_path)
```

```python
DATABASE['cmeS'].head()
```

```python
DATABASE['cmeS'].DateOfSale = DATABASE['cmeS'].DateOfSale.apply(lambda x: datetime.datetime.strptime(x, "%m/%d/%Y"))
```

```python
df1 = DATABASE['cmeS'].set_index('DateOfSale')
```

```python
df1_new = df1.resample('M').price.mean()
df1_new.pad().plot(figsize =(16,9))
df1_new.bfill().plot()
df1_new.interpolate().plot()
plt.legend(['forward fill','backward fill','interpolate'])
```
